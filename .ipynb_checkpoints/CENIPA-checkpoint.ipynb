# 1. Motivação 

Os recentes acidentes aéreos veículados na TV e redes sociais despertaram a curiosidade sobre o assunto, acompanhando o canal do Youtube do Lito Sousa, [Aviões e Músicas](https://www.youtube.com/channel/UCLXl1V6n82Dyg1VhVgSL0nw), tomei conhecimento da existência do Centro de Investigação e Prevenção de Acidentes Aeronáuticos (CENIPA), o CENIPA é o órgão do Comando da Aeronáutica responsável pelas atividades de investigação e prevenção de acidentes aeronáuticos.

Ao acessar o [portal do CENIPA](https://www2.fab.mil.br/cenipa/) encontrei as bases de dados necessárias para entender os fatatores e cenários de ocorrências aeronáuticas.

```python
import os
import pandas as pd


path_dataset = 'datasets'
if not os.path.exists(path_dataset):
    os.makedirs(path_dataset)

filename_df_ocorrencia = f'{path_dataset}/ocorrencia.xlsx'
filename_df_ocorrencia_tipo = f'{path_dataset}/ocorrencia_tipo.xlsx'
filename_df_aeronave = f'{path_dataset}/aeronave.xlsx'
filename_df_fator_contribuinte = f'{path_dataset}/fator_contribuinte.xlsx'
filename_df_recomendacao = f'{path_dataset}/recomendacao.xlsx'
filename_df_relatorio = f'{path_dataset}/relatorio.xlsx'

urls_df = {
    filename_df_ocorrencia: "http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia.csv",
    filename_df_ocorrencia_tipo: "http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia_tipo.csv",
    filename_df_aeronave: "http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/aeronave.csv",
    filename_df_fator_contribuinte: "http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/fator_contribuinte.csv",
    filename_df_recomendacao: "http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/recomendacao.csv"
}


for filename,url in urls_df.items():
    pd.read_csv(url, sep=';').to_excel(filename, index=False)
```

# 3. Tratamento dos Dados

As 5 bases de dados do CENIPA disponibilizados [Portal Brasileiro de Dados Abertos](https://dados.gov.br/dataset/ocorrencias-aeronauticas-da-aviacao-civil-brasileira) estão relacionadas pelo código da ocorrência.

obs: Foi realizado a alteração no nome da coluna das bases RECOMENDACAO, OCORRENCIA_TIPO, FATOR_CONTRIBUINTE e AERONAVE para *codigo_ocorrencia*, sem número no final, assim facilitando a manipulação/cruzamento dos dados.

```python
# Função para retornar a porcentagem de missing values de um dataset
def check_missing_values(df):
    df = round((df.isnull().sum()/df.shape[0])*100,2) 
    df = df[df>0]   
    return df
```

# 4. Análise de Dados

Os cruzamentos de dados foram realizados utilizando querys SQL's no Pandas.

```python
import sqlite3
con = sqlite3.connect(':memory:')

df_ocorrencia.to_sql(name='ocorrencia', con=con, index=False)
df_ocorrencia_tipo.to_sql(name='ocorrencia_tipo', con=con, index=False)
df_aeronave.to_sql(name='aeronave', con=con, index=False)
df_fator_contribuinte.to_sql(name='fator_contribuinte', con=con, index=False)
df_recomendacao.to_sql(name='recomendacao', con=con, index=False)
```

Para visualização dos dados foi utilizado a bibilioteca [Seaborn](https://seaborn.pydata.org/).

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_palette( sns.color_palette("hls", 1) )
order_classificacao = ["INCIDENTE", "INCIDENTE GRAVE", "ACIDENTE"]
```

## 4.1 Ocorrências nos últimos 10 anos

```python
query1 = '''
        SELECT ocorrencia_classificacao
              ,COUNT(*) as ocorrencia_quantidade
        FROM ocorrencia
        GROUP BY ocorrencia_classificacao
        ORDER BY ocorrencia_quantidade DESC
        '''

query2 = ''' 
        SELECT ocorrencia_ano
               ,ocorrencia_classificacao
               ,COUNT(*) as ocorrencia_quantidade
        FROM ocorrencia
        GROUP BY ocorrencia_ano
                ,ocorrencia_classificacao
        '''

df1 = pd.read_sql_query(query1, con)
df2 = pd.read_sql_query(query2, con)
ocorrencias = df2['ocorrencia_quantidade'].sum()

fig, axes =  plt.subplots(ncols=2, figsize=(20,5))

g1 = sns.barplot(ax=axes[0],x='ocorrencia_classificacao', y='ocorrencia_quantidade', data=df1, order=order_classificacao  )
g1.set(title = 'Ocorrências por Classificação', ylabel='Quantidade de Ocorrências', xlabel='Classificação')

g2 = sns.lineplot(ax=axes[1], x='ocorrencia_ano', y='ocorrencia_quantidade', hue='ocorrencia_classificacao',
                   hue_order =order_classificacao,  data=df2)
g2.set(title=f'Panorama das {ocorrencias} ocorrências nos últimos 10 anos', xlabel='Ano', ylabel='Quantidade de Ocorrências')
g2.legend(title='Classificação')

```

Nos últimos 10 anos o CENIPA registrou 6114 ocorrências, mais da metade das ocorrências, cerca de 3393 ocorrências, são de incidentes aeronauticos, 1930 acidentes e 791 incidentes graves.

Para ocorrências classificadas como incidente o ano de 2013 lidera com 386 incidentes registrados, é o observado uma diminuição no número de ocorrências desse tipo após 2013.  

Para ocorrências classificadas como incidentes graves o ano de 2018 lidera com 81 ocorrências deste tipo, seguido dos anos de 2019 e 2020, com 80 e 77 incidentes graves, respectivamente.

Para ocorrências classificadas como acidentes o ano de 2012 lidera com 206 acidentendes, seguido do ano de 2013 com 199 acidentes e 2011 com 187 acidentes. 

## 4.2 Fatalidades nos últimos 10 anos

```python
query = '''
        SELECT ocorrencia.ocorrencia_ano
              ,SUM(aeronave.aeronave_fatalidades_total) AS fatalidades
        FROM ocorrencia INNER JOIN aeronave
        ON ocorrencia.codigo_ocorrencia = aeronave.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_ano
        '''
df = pd.read_sql_query(query, con)
fatalidades = df['fatalidades'].sum()

fig, axes =  plt.subplots(ncols=1, figsize=(20,5))
g = sns.lineplot(ax=axes, x='ocorrencia_ano', y='fatalidades', marker='o', data=df, )
g.set(title=f'Panorama das {fatalidades} fatalidades nos últimos 10 anos', xlabel='Ano', ylabel='Quantidade de Fatalidades')
```

Os anos de 2011 e 2012 lideram o ranking de fatalidades aeronauticas regsistradas nos ultimos 10 anos pelo CENIPA, foram 110 óbitos. Após 2012 é possível observar uma tendência decrescente na quantidade de óbitos tendo apenas 2 picos de aumento nos anos de 2016 e 2018, com 104 e 80 óbitos, respectivamente.

## 4.3 Tipo de ocorrência por classificação da ocorrência

```python
query = '''
        SELECT ocorrencia.ocorrencia_classificacao
              ,ocorrencia_tipo.ocorrencia_tipo
              ,COUNT(*) as ocorrencia_quantidade
        FROM ocorrencia INNER JOIN ocorrencia_tipo
        ON ocorrencia.codigo_ocorrencia = ocorrencia_tipo.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_classificacao
                ,ocorrencia_tipo.ocorrencia_tipo
        ORDER BY ocorrencia_quantidade DESC
        '''
df = pd.read_sql_query(query, con)
df = df.groupby(by='ocorrencia_classificacao').head(10)

g = sns.catplot(y='ocorrencia_tipo', x='ocorrencia_quantidade', col='ocorrencia_classificacao',
                col_order=order_classificacao, data=df, kind='bar', ci=None)
g.set(ylabel = 'Tipo de Ocorrência', xlabel='Quantidade de Ocorrências', )
g.set_titles("{col_name}")
```

Para ocorrências classficadas como incidentes, o top 3 de ocorrências mais registradas pelo CENIPA são:
 1. Estouro de Pneu (635)
 2. Falha ou mau funcionamento de Sistema/Componente (604)
 3. Colisão com Ave (293)
 
Para ocorrências classificadas como incidentes graves, o top 3 de ocorrências mais registradas pelo CENIPA são:
1. Perda de controle no solo (132)
2. Com trem de pouso (131)
3. Falha do motor em voo (115)

Para ocorrências classificadas como acidentes, o top 3 de ocorrências mais registradas pelo CENIPA são:
1. Falha do motor em voo (384)

3. Perda de controle no solo (384)

## 4.5 Ocorrência por tipo de operação da aeronave

```python
query = '''
        SELECT ocorrencia.ocorrencia_classificacao
              ,aeronave.aeronave_tipo_operacao
              ,COUNT(*) AS ocorrencia_quantidade
        FROM ocorrencia INNER JOIN aeronave
        ON ocorrencia.codigo_ocorrencia = aeronave.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_classificacao
                ,aeronave.aeronave_tipo_operacao
        ORDER BY aeronave.aeronave_tipo_operacao DESC
                 ,ocorrencia_quantidade DESC
 
        '''
df = pd.read_sql_query(query, con)
g = sns.catplot(y='aeronave_tipo_operacao', x='ocorrencia_quantidade', col='ocorrencia_classificacao',
                col_order=order_classificacao, data=df, kind='bar', ci=None)
g.set(ylabel = 'Tipo de Operação da Aeronave', xlabel='Quantidade de Ocorrências', )
g.set_titles("{col_name}")
```

Para ocorrências classificadas como incidente temos que o top 3 é composto por aeronaves do tipo de operação regular com 1198 registros, privada com 845 registros e Táxi Aéreo com 572 registros.

O cenário de ocorrências de incidentes graves temos que aeronaves do tipo privada lideram com cerca de 318 registros, seguidos de aeronaves de instruções com 160 registros e táxi aéreo com 98 registros.

As aeronaves de tipo de operação privada lideram o top 3 de acidentes, são cerca de 855 registros de acidentes nessa modalidade, mais que o dobro do segundo colocado, aeronaves agrícolas com 412 acidentes, seguido de aeronaves de instrução com 243 registros.

## 4.6 Ocorrências por quantidade de motores da aeronave

```python
query = '''
        SELECT ocorrencia.ocorrencia_classificacao
              ,aeronave.aeronave_motor_quantidade
              ,COUNT(*) AS ocorrencia_quantidade
        FROM ocorrencia INNER JOIN aeronave
        ON ocorrencia.codigo_ocorrencia = aeronave.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_classificacao
                ,aeronave.aeronave_motor_quantidade
        ORDER BY ocorrencia_quantidade DESC
        '''
df = pd.read_sql_query(query, con)

g = sns.catplot(y='aeronave_motor_quantidade', x='ocorrencia_quantidade', col='ocorrencia_classificacao',
                order = ['MONOMOTOR', 'BIMOTOR','TRIMOTOR','QUADRIMOTOR', 'SEM TRAÇÃO', '***'],  col_order=order_classificacao,  data=df, kind='bar', ci=None)
g.set(ylabel = 'Quantidade de Motores', xlabel='Quantidade de Ocorrências', )
g.set_titles("{col_name}")
```

Em ocorrências classificadas como incidentes as aeronaves com dois motores apresentam o maior número de ocorrências, quando o cenário é ocorrencias classificadas como incidentes graves ou acidentes as aeronaves com apenas um motor ficam no top 1 das ocorrências, acidentes com monomotores chegam a 80% das ocorrências desse cenário.

## 4.7 Tipo de ocorrência por quantidade de motores da aeronave

```python
query = '''
        SELECT ocorrencia.ocorrencia_classificacao
              ,ocorrencia_tipo.ocorrencia_tipo
              ,aeronave.aeronave_motor_quantidade
              ,COUNT(*) as ocorrencia_quantidade
        FROM ocorrencia INNER JOIN ocorrencia_tipo
        ON ocorrencia.codigo_ocorrencia = ocorrencia_tipo.codigo_ocorrencia
        INNER JOIN aeronave
        ON ocorrencia.codigo_ocorrencia = aeronave.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_classificacao
                ,ocorrencia_tipo.ocorrencia_tipo
                ,aeronave.aeronave_motor_quantidade
        ORDER BY ocorrencia_quantidade DESC
        '''
df = pd.read_sql_query(query, con)
df = df.head(50)
g = sns.catplot(y='ocorrencia_tipo', x='ocorrencia_quantidade', 
                col='ocorrencia_classificacao', col_order=order_classificacao,
                row='aeronave_motor_quantidade', row_order = ['MONOMOTOR','BIMOTOR'],
                data=df, kind='bar', ci=None)
g.set(ylabel = 'Tipo de Ocorrência', xlabel='Quantidade de Ocorrências', )
g.set_titles("{col_name} COM {row_name}")
```

Em acidentes com aeronaves monomotores a falha do motor em voo é a principal ocorrência registrada, seguido de perda de controle em voo.

Falha ou mau funcionamento de sistema/componente é o principal incidente registrado em bimotores, seguido de estouro de pneu e colisão com ave.

## 4.9 Principais Fatores Contribuíntes das Ocorrências

Centro de Investigação e Prevenção de Acidentes Aeronáuticos (CENIPA) não trabalha com "causa" de acidente, mas com fatores contribuintes. "Causa" se refere a um fator que se sobressai, que seja preponderante, e a investigação do CENIPA não elege um fator como o principal. Ao contrário, trabalha com uma série de fatores contribuintes que possuem o mesmo grau de influência para a culminância do acidente.

Todo acidente resulta de uma seqüência de acontecimentos e nunca de um fator isolado, os fatores contribuintes são dados por:

1. FATOR HUMANO: Se refere à complexidade biológica do ser humano em seus aspectos fisiológicos e psicológicos.

2. FATOR MATERIAL: Se refere à aeronave em seus aspectos de projeto de fabricação e manuseio e estoque do material.

3. FATOR OPERACIONAL: Se refere ao desempenho do ser humano nas ações e atividades relacionadas com o vôo. 

```python
query = '''
        SELECT ocorrencia.ocorrencia_classificacao
              ,fator_contribuinte.fator_area
              ,COUNT(*) AS ocorrencia_quantidade
        FROM ocorrencia INNER JOIN fator_contribuinte
        ON ocorrencia.codigo_ocorrencia = fator_contribuinte.codigo_ocorrencia
        GROUP BY ocorrencia.ocorrencia_classificacao
                ,fator_contribuinte.fator_area
        ORDER BY ocorrencia_quantidade DESC
        '''
df = pd.read_sql_query(query, con)
df = df.groupby('ocorrencia_classificacao').head(5)
g = sns.catplot(y='fator_area', x='ocorrencia_quantidade', col='ocorrencia_classificacao',
                 col_order=order_classificacao, data=df, kind='bar', ci=None)
g.set(ylabel = 'Tipo de Motor', xlabel='Quantidade de Ocorrências', )
g.set_titles("{col_name}")
```

## 4.10 Divulgação de Relatórios das Ocorrências

```python
query = '''
         SELECT  divulgacao_relatorio_publicado AS "Divulgado"
                 ,COUNT(*) AS "Quantidade"
                 ,ROUND( COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS "%"
         FROM ocorrencia
         GROUP BY Divulgado
         '''
df = pd.read_sql_query(query, con)
df
```

Cerca de 74% das ocorrências dos CENIPA ainda não tiveram seus relatórios liberados.

```python

query1 = '''
         SELECT  ocorrencia_classificacao
                 ,divulgacao_relatorio_publicado
                 ,COUNT(*) AS ocorrencia_quantidade
         FROM ocorrencia
         GROUP BY divulgacao_relatorio_publicado
                  ,ocorrencia_classificacao
         '''


query2 = '''
        SELECT ocorrencia_classificacao
              ,divulgacao_anos
        FROM ocorrencia 
        '''
df1 = pd.read_sql_query(query1, con)
df2 = pd.read_sql_query(query2, con)

fig, axes =  plt.subplots(ncols=2, figsize=(20,5))


g1 = sns.barplot(ax=axes[0],x='divulgacao_relatorio_publicado', y='ocorrencia_quantidade', 
                 hue='ocorrencia_classificacao', hue_order =order_classificacao, order = ['SIM','NÃO'], data=df1, ci=None  )
g1.set(title = 'Divulgação do Relatório da Ocorrência', ylabel='Quantidade de Ocorrências', xlabel='Classificação')
g1.legend(title='Classificação')


g2 = sns.boxplot(ax=axes[1], y="divulgacao_anos", x='ocorrencia_classificacao', order=order_classificacao, data=df2)
g2.set(title='Anos até a divulgação do relatório', xlabel='Classificação', ylabel='Anos', )

```

```python
df2.groupby('ocorrencia_classificacao').mean()
```

74% das ocorrências que tem relatório liberado são de acidentes, em média o CENIPA divulga o relatório de acidentes em 2 anos. Incidentes graves são as ocorrências que mais apresentam variação de anos na divulgação do relatório.

## 4.14 Recomendações das Ocorrências 

```python

query1 = '''
         SELECT  recomendacao_status
                 ,COUNT(*) AS ocorrencia_quantidade
         FROM recomendacao
         GROUP BY recomendacao_status
         '''

query2 = '''
         SELECT  recomendacao_destinatario_sigla
                 ,COUNT(*) AS ocorrencia_quantidade
         FROM recomendacao
         GROUP BY recomendacao_destinatario_sigla
         ORDER BY ocorrencia_quantidade DESC
         '''
df1 = pd.read_sql_query(query1, con)
df2 = pd.read_sql_query(query2, con)
df2=df2.head(10)

fig, axes =  plt.subplots(ncols=2, figsize=(20,5))
g1 = sns.barplot(ax=axes[0], y='recomendacao_status', x='ocorrencia_quantidade', data=df1, ci=None  )
g1.set(title = 'Status das Recomendações', xlabel='Quantidade de Ocorrências', ylabel='Status')


g2 = sns.barplot(ax=axes[1], y='recomendacao_destinatario_sigla', x='ocorrencia_quantidade', data=df2, ci=None  )
g2.set(title = 'Destinatário das Recomendações', xlabel='Quantidade de Ocorrências', ylabel='Sigla Destinatário')
```

Em relação as recomendações, 61% das recomendações do CENIPA são cumpridas, a [ANAC](https://www.gov.br/anac/pt-br) é o orgão que mais recebe essas recomendações, cerca de 91% das recomendações do CENIPA.
